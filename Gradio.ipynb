```python
import torch
import gradio as gr
from transformers import pipeline

device = 0 if torch.cuda.is_available() else -1
clf = pipeline("image-classification", model="google/vit-base-patch16-224", device=device)
```

```python
def predict(image, top_k):
    # returns top-k labels for an uploaded image
    return clf(image, top_k=int(top_k))
```

```python
# Gradio Layout: Rows + Columns + Image + Slider

with gr.Blocks() as demo:
    gr.Markdown("# Minimal Gradio + Transformers Demo\nUpload an image â†’ get top-k predictions.")

    with gr.Row():
        with gr.Column():
            img = gr.Image(type="pil", label="Image")
            k = gr.Slider(1, 10, value=5, step=1, label="Top-K")
            btn = gr.Button("Run")

        with gr.Column():
            out = gr.Label(label="Predictions")

    btn.click(predict, inputs=[img, k], outputs=out)

demo
```
